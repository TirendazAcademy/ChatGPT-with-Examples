`Write an SQLite query in Python to create a database. Write an SQLite query in Python to create a database.`

```python
import sqlite3

# Connect to the database
conn = sqlite3.connect("mydatabase.db")

# Create a cursor
cursor = conn.cursor()

# Create the table
cursor.execute("CREATE TABLE demo (id INTEGER PRIMARY KEY, name TEXT, age INTEGER, job TEXT)")

# Commit the transaction
conn.commit()

# Close the connection
conn.close()
```

`Add 100 random values inside this table.`

```python
import sqlite3
import random

# Connect to the database
conn = sqlite3.connect("mydatabase.db")

# Create a cursor
cursor = conn.cursor()

# Generate 100 random values
for i in range(100):
    name = "Name" + str(i)
    age = random.randint(18, 60)
    job = "Job" + str(i)

    # Insert the values into the table
    cursor.execute("INSERT INTO demo (name, age, job) VALUES (?, ?, ?)", (name, age, job))

# Commit the transaction
conn.commit()

# Close the connection
conn.close()
```

` Write a SQL query in Python to show the first 5 values in this table.`

```python
import sqlite3

# Connect to the database
conn = sqlite3.connect("mydatabase.db")

# Create a cursor
cursor = conn.cursor()

# Execute the SELECT query
cursor.execute("SELECT * FROM demo LIMIT 5")

# Fetch the results
results = cursor.fetchall()

# Print the results
for row in results:
    print(row)

# Close the connection
conn.close()
```

`Write a SQL query in Python to see the average of the age column in this table.`

```python
import sqlite3

# Connect to the database
conn = sqlite3.connect("mydatabase.db")

# Create a cursor
cursor = conn.cursor()

# Execute the SELECT query
cursor.execute("SELECT AVG(age) FROM demo")

# Fetch the result
result = cursor.fetchone()

# Print the result
print(result[0])

# Close the connection
conn.close()
```

`Write a SQL query in Python to see the names of the tables in the database named my_demo.db.`

```python
import sqlite3

# Connect to the database
conn = sqlite3.connect("my_demo.db")

# Create a cursor
cursor = conn.cursor()

# Execute the SELECT query
cursor.execute("SELECT name FROM sqlite_master WHERE type='table'")

# Fetch the results
results = cursor.fetchall()

# Print the results
for row in results:
    print(row[0])

# Close the connection
conn.close()
```

`Write a SQL query in Python to see the values less than 5000 in the sal column in the emp table.`

```python
import sqlite3

# Connect to the database
conn = sqlite3.connect("my_demo.db")

# Create a cursor
cursor = conn.cursor()

# Execute the SELECT query
cursor.execute("SELECT * FROM emp WHERE sal < 5000")

# Fetch the results
results = cursor.fetchall()

# Print the results
for row in results:
    print(row)

# Close the connection
conn.close()
```
